```python
def hex_to_binary(s):
    """convert hex string to binary string"""
    assert len(s) == 32
    return format(int(s, 16), '0>0128b')
```

```python
def binary_to_maze(s):
    return s.replace('1', '#').replace('0', '.')
```

```python
def count_zeros(s):
    return sum(map(int, list(s)))
assert count_zeros('011000110001010101') == 8
```

```python
input = 'stpzcrnm'
maze = []
for i in range(128):
    s = input + '-' + str(i)
    h = knot_hash(list(s))
    b = hex_to_binary(h)
    assert len(b) == 128
    maze.append(binary_to_maze(b))
```

```python
len(maze)
```

```python
# 2016 day 13
def is_valid(r, c):
    if r >= 0 and c >= 0 and r < 128 and c < 128:
        return maze[r][c] == '#'

def generate_moves(pos):
    r, c = pos
    new_positions = [(r+1, c), (r-1, c), (r, c+1), (r, c-1)]
    return [newpos for newpos in new_positions if is_valid(*newpos)]


def bfs_color_group(r, c):
    """breath first search"""
    visited, queue = set(), [(r, c)]
    while queue:
        pos = queue.pop(0)
        r, c = pos
        #maze[r][c] = 'X'
        if pos not in visited:
            visited.add(pos)
            queue.extend(generate_moves(pos))
    return visited
```

```python
visited = set()
n = 0
for r in range(128):
    for c in range(128):
        if maze[r][c] == '.':
            continue
        if (r,c) in visited:
            continue
        group = bfs_color_group(r, c)
        print('at %d %d found group of size %d' % (r, c, len(group)))
        visited.update(group)
        n += 1
```
